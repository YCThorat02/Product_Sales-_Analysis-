```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.strftime('%b')  
df['Month_Num'] = df['Date'].dt.month       

monthly_sales = df.groupby(['Year', 'Month', 'Month_Num'])['Sales'].sum().reset_index()
monthly_sales = monthly_sales.sort_values(by='Month_Num')
heatmap_data = monthly_sales.pivot(index='Year', columns='Month', values='Sales')

plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_data, annot=True, fmt=".0f", cmap='Blues', linewidths=0.5)
plt.title("Month vs Year ")
plt.xlabel("Month")
plt.ylabel("Year")
plt.tight_layout()
plt.show()
```
